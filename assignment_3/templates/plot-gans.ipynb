```python
import numpy as np
# grads_w
# textgen_inputlength_30_hiddenunits_128_lr_0.01_batchsize_64_1574712613.csv
import os
results_path = "part1/results"
for f in os.listdir(results_path):
    if  "RNN_grad_hidden_state_seq.txt" in f:
        print(f)
        data = pd.read_csv(os.path.join(results_path, f), header=None)
        plt.plot(data[0], np.log(data[1]), label="vanillaRNN")

for f in os.listdir(results_path):
    if  "LSTM_grad_hidden_state_seq.txt" in f:
        print(f)
        data = pd.read_csv(os.path.join(results_path, f), header=None)
        plt.plot(data[0], data[1], label="LSTM")
plt.legend()
plt.xlabel("steps")
plt.ylabel("gradients of hidden state")
# plt.title("input_length=40, hidden_units=128, lr=0.01")
plt.yscale('symlog')

plt.show()
```
